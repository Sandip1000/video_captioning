```python
from huggingface_hub import hf_hub_download

# example
file_path = hf_hub_download(
    repo_id="Sandip1000/vatex_nepali",   # change this
    repo_type="dataset",
    filename="VATEX_ne.zip",
    local_dir=fr"D:\test" # exact zip filename
)

print("Downloaded to:", file_path)
```

```python
import os
import zipfile
# unzip
extract_path = fr"D:\test"
zip_path = fr"D:\test\VATEX_ne.zip"
os.makedirs(extract_path, exist_ok=True)

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

print("Downloaded zip:", zip_path)
print("Unzipped folder:", extract_path)
```
